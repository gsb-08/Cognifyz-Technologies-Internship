Task: Feature Engineering

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv("Dataset.csv")
dataset.head()
```

Extract additional features from the existing columns, such as the length of the restaurant name or address.

```python
dataset["Restaurant Name Length"]=dataset["Restaurant Name"].apply(lambda x:len(str(x)))
dataset["Address Length"]=dataset["Address"].apply(lambda x:len(str(x)))
```

```python
dataset[["Restaurant Name","Restaurant Name Length","Address","Address Length"]]
```

Create new features like "Has Table Booking"or "Has Online Delivery" by encoding
categorical variables.

```python
dataset["Has Table Booking"]=dataset["Has Table booking"].apply(lambda x:1 if x=="Yes"else 0)
dataset["Has Online Delivery"]=dataset["Has Online delivery"].apply(lambda x:1 if x=="Yes"else 0)
```
