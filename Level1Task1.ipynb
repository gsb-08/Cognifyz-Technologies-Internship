Task: Data Exploration and Preprocessing

Explore the dataset and identify the number
of rows and columns.

Check for missing values in each column and
handle them accordingly.

Perform data type conversion if necessary.
Analyze the distribution of the target variable
("Aggregate rating") and identify any class
imbalances.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv("Dataset.csv")
dataset.head()
```

```python
# Get the number of rows and columns
num_rows, num_columns = dataset.shape
print(f"The dataset has {num_rows} rows and {num_columns} columns.")
```

```python
# Check for missing values in each column
missing_values = dataset.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
# Handling missing values
dataset["Cuisines"].fillna("not specified",inplace=True)
```

```python
dataset.isnull().sum()
```

```python
# Perform data type conversion if necessary
# Example: Converting "Aggregate rating" to numeric if it's not already
if dataset['Aggregate rating'].dtype != 'float64':
    dataset['Aggregate rating'] = pd.to_numeric(dataset['Aggregate rating'], errors='coerce')
dataset.head()
```

```python
# Analyze the distribution of the target variable
print("Aggregate rating distribution:")
print(dataset['Aggregate rating'].value_counts())
```

```python
# Plot the distribution
plt.figure(figsize=(8, 6))
dataset['Aggregate rating'].hist(bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Identify class imbalances
class_counts = dataset['Aggregate rating'].value_counts()
print("\nClass Imbalance:")
print(class_counts)
```

```python
# Visualize class imbalance
plt.figure(figsize=(8, 6))
class_counts.plot(kind='bar', color='red', edgecolor='black')
plt.title('Class Imbalance in Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```
