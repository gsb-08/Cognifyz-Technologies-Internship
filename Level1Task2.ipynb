Task2:Descriptive Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
dataset = pd.read_csv("Dataset.csv")
dataset.head()
```

Calculate basic statistical measures (mean,median, standard deviation, etc.) for numerical columns.

```python
dataset.describe()
```

Explore the distribution of categorical variables like "Country Code", "City" and "Cuisines".

```python
import seaborn as sns;sns.set(color_codes=True)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x="Country Code",data=dataset,palette='cividis')
plt.title('Distribution of Restaurant by Country Code')
plt.xlabel('Country Code')
plt.ylabel('No. of Restaurant')
plt.grid(True)
plt.show()
```

Identify the top cuisines and cities with the highest number of restaurants.

```python
top_countries=dataset=dataset['Country Code'].value_counts().head()
print("Top 5 countries with the highest no. of Restarant : ")
print(top_countries)
```

```python
plt.figure(figsize=(14, 6))
sns.countplot(x="City",data=dataset, order=dataset["City"].value_counts().head(20).index,palette='Set2')
plt.title('Distribution of Restaurant by City:')
plt.xlabel('City')
plt.ylabel('No. of Restaurant')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 6))
cuisines_count=dataset['Cuisines'].value_counts()
cuisines_count.head(20).plot(kind="bar",color=sns.color_palette("Set2"))
plt.title('Top 20 Cuisines with the highest no. of Restarant :')
plt.xlabel('Cuisines')
plt.ylabel('No. of Restaurant')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
top_cities=dataset["City"].value_counts().head(10)
print("Top 10 cities with the highest no. of Restarant : ")
print(top_cities)
```

```python
top_cusines=cuisines_count.head(10)
print("Top 10 cusines with the highest no. of Restarant : ")
print(top_cusines)
```
